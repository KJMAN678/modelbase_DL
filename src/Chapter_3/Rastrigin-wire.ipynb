```julia
using LinearAlgebra
using Plots
gr()
using Random
Random.seed!(1)
```

```julia
n = 2
A = 10
function f(x)
    return A*n + sum(x.^2 - A*cos.(2pi*x))
end
```

```julia
x = -2:0.1:2
y = -2:0.1:2

z = [Float64(f([xi,yi])) for xi in x, yi in y];
```

```julia
wireframe(x,y,z, color=:black)
```
